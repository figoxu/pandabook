```python
import pandas as pd
path = "./data/高考.csv" 
exame = pd.read_csv(path)
exame["高考人数"] = exame["高考人数 "].str.replace("万", "")
exame["高考人数"] = exame["高考人数"].astype("int")
exame["录取率"] = exame["录取率"].str.replace("%", "")
exame["录取率"] = exame["录取率"].astype("float")


exame["录取人数"] = exame["录取人数"].str.replace("万", "")
exame["录取人数"] = exame["录取人数"].astype("float")
exame["录取人数"] = exame["录取人数"].astype("int")

exame = exame.sort_values(by=['年份'],ascending=True)
exame
```

```python
from pyecharts import options as opts
from pyecharts.charts import Bar, Line

x_data = exame["年份"].apply(str).tolist()
bar = (
    Bar(init_opts=opts.InitOpts(width="1000px", height="400px"))
    .add_xaxis(
       x_data
    )
    .set_colors(["#a1c84f","#42b0cd","#2c4d75"])
    .add_yaxis(series_name="高考人数(单位: 万人)",yaxis_data= exame["高考人数"].tolist(),
 yaxis_index=0, color="#a1c84f")
    .add_yaxis("录取人数(单位: 万人)", exame["录取人数"].tolist(), color="#42b0cd")
    .extend_axis(
        yaxis=opts.AxisOpts(
            name="录取率",
            type_="value",
            min_=0,
            max_=100,
            interval=10,
            axislabel_opts=opts.LabelOpts(formatter="{value} %"),
        )
    )
    .set_global_opts(
        tooltip_opts=opts.TooltipOpts(
            is_show=True, trigger="axis", axis_pointer_type="cross"
        ),
        xaxis_opts=opts.AxisOpts(
            type_="category",
            axispointer_opts=opts.AxisPointerOpts(is_show=True, type_="shadow"),
        ),
        yaxis_opts=opts.AxisOpts(
            name="人数",
            type_="value",
            min_=500,
            max_=1400,
            interval=100,
            axislabel_opts=opts.LabelOpts(formatter="{value} 万人"),
            axistick_opts=opts.AxisTickOpts(is_show=True),
            splitline_opts=opts.SplitLineOpts(is_show=True),
        ),
    )
)


x_data = exame["年份"].apply(str).tolist()
y_data = exame["录取率"].tolist()
line = (
    Line()
    .add_xaxis(x_data)
    .set_colors(["#2c4d75"])
    .add_yaxis(series_name="录取率", 
        yaxis_index=1,
               y_axis=exame["录取率"], color="#2c4d75",
              label_opts=opts.LabelOpts(is_show=True),)
    
)

bar.overlap(line).render_notebook()

# charts相关API参考 https://gallery.pyecharts.org/#/Line/multiple_x_axes
```
